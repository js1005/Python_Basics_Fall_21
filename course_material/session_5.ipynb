## Univariate plots

You might be interested in the characteristics of a single variable (univariate analysis/plots) or in the relationship between two or more variables (multivariate analysis/plots). We begin with univariate plots, i.e., we visualize the characteristics of single variables.

### 1. Distribution plots/histograms

As the name suggests, distribution plots visualize the distribution of the values of a single variable: E.g., what is the distribution of the 'Survived' variable, i.e., how many '1' values (= Survived) and '0' values (= Died) are there?

```python
plt.hist(list(df_titanic['Survived']))
```

```python
plt.hist(list(df_titanic['Survived']), color='r')
```

```python
plt.hist(list(df_titanic['Survived']), color='r', edgecolor='black')
```

```python
plt.hist(list(df_titanic['Survived']))
plt.title('Number of passengers who died (=0) versus survived (=1)')
```

```python
plt.hist(list(df_titanic['Survived']))
plt.title('Number of passengers who died (=0) versus survived (=1)')
plt.xlabel('Survived')
```

```python
plt.hist(list(df_titanic['Survived']))
plt.title('Number of passengers who died (=0) versus survived (=1)')
plt.xlabel('Survived')
plt.ylabel('# of passengers')
```

```python
plt.hist(list(df_titanic['Survived']), bins=2)
```

## 2. Multivariate plots

Next, we visualize the relationship between two variables. That is, we are interested in how the values of a variable y vary with the values of another variable x. In the Titanic context, we might be interested in the following questions: What is the relationship between survival chances and gender? Are women more likely to survive than men because they were given priority on the lifeboats? Or were men more likely to survive than women because they had greater physical strength? We might also be interested in the relationship between 'Age' and 'Fare': Were pricy tickets reserved to older passengers?

Whenever we are interested in the relationship between two variables and we would like to visualize it, we first have to consider the SCALE of variables or, in the Python language, the datatype. If we study the relationship between two categorical variables, like gender and survival, we will choose different kinds of plots than if we study the relationship between one categorical and one numeric or even between two numeric variables.

Let us consider the relationship between two categorical variables first: gender and survival.

```python
sns.catplot(x='Sex', y='Survived', data=df_titanic, kind='bar')
```

```python
sns.catplot(x='Sex', y='Survived', data=df_titanic, kind='point')
```

```python
sns.catplot(x='Sex', y='Survived', data=df_titanic, kind='violin')
```

Females clearly have a survival advantage!

Next, let us consider the relationship between a categorical and numeric variable, namely 'Survived' and 'Fare': Were passengers with pricy tickets more likely to survive, e.g., because they were given preferential treatment on lifeboats?

```python
sns.catplot(x='Survived', y='Fare', data=df_titanic)
```

```python
sns.catplot(x='Survived', y='Fare', data=df_titanic, kind='swarm')
```

We can extend our visualizations by another, third variable/dimension, e.g., 'Pclass' (Passenger class). We color the observations based on 'Pclass'.

```python
sns.catplot(x='Survived', y='Fare', hue='Pclass', data=df_titanic, kind='swarm')
```

The boxplot adds quartiles and median ...

```python
sns.catplot(x='Survived', y='Fare', data=df_titanic, kind='box')
```

Finally, we might be interested in the relationship between two numeric variables: What is the relationship between 'Age' and 'Fare'? Were older passengers more likely to buy pricy tickets?

We can visualize this relationship with a scatterplot where each scatter plot corresponds to one observation.

```python
sns.scatterplot(x='Age', y='Fare', data=df_titanic)
```

We cannot see much because the data is quite crowded for low fares. Let's zoom in a bit:

```python
sns.scatterplot(x='Age', y='Fare', data=df_titanic)
plt.ylim(0, 50)
```

There seems to be no clear relationship between 'Age' and 'Fare'. For illustration purposes, let's nevertheless regress 'Age' on 'Fare':

```python
sns.lmplot(x='Age', y='Fare', data=df_titanic)
```

```python
# remove 'Fare' outliers
sns.lmplot(x='Age', y='Fare', data=df_titanic.loc[df_titanic['Fare'] < df_titanic['Fare'].quantile(.95)])
```

```python
sns.lmplot(x='Age', y='Fare', hue='Pclass', data=df_titanic.loc[df_titanic['Fare'] < df_titanic['Fare'].quantile(.95)])
```

Within the passenger classes, 'Fare' seems to decrease with 'Age' - passengers' age alone does not seem to be a very good predictor of 'Fare'.
